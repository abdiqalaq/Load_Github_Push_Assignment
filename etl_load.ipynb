```python
import pandas as pd
import sqlite3
from pathlib import Path

# --- Load transformed CSVs ---
df_full = pd.read_csv('transformed/transformed_full.csv')
df_incr = pd.read_csv('transformed/transformed_incremental.csv')

# --- Create loaded directory if it doesn't exist ---
Path("loaded").mkdir(exist_ok=True)

# --- Option 1: Load into SQLite ---
conn = sqlite3.connect('loaded/full_data.db')
df_full.to_sql('full_data', conn, if_exists='replace', index=False)
df_incr.to_sql('incremental_data', conn, if_exists='replace', index=False)

# --- Verify load ---
pd.read_sql('SELECT * FROM full_data LIMIT 5', conn)
```
